```python
df = pd.read_csv('symbol_info_3-25.csv')
```

```python
df_sample = df.sample(30, random_state=24)
```

```python
df_sample = df_sample[['industry', 'enterprise_value', 'current_price']]
```

```python
sample_tot_value= df_sample['enterprise_value'].sum()
```

```python
df_sample['persentage_value']= (df_sample['enterprise_value']/sample_tot_value)*100
```
